## Grep
> searches any given input files, selecting lines that
     match one or more patterns

To learn more visit:

* https://www.gnu.org/software/grep/manual/grep.html
* https://www.computerhope.com/unix/ugrep.htm

```python
!man grep
```

```python
!ls | grep .txt
```

```python
!ls | grep \\.txt\$
```

#### Ignore Case

```python
!grep "hello" Folder_1/hello
```

```python
!grep -i "hello" Folder_1/hello
```

#### Search multiple files

```python
!grep -i "hello" Folder_1/*
```

```python
!grep -i -r "hello" Folder_1  #recursively search subdirectories
```
